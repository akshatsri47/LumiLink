```python
# # Install Unsloth library
# !pip install unsloth

# # Install additional dependencies
# !pip install transformers accelerate bitsandbytes sentencepiece

# Import required libraries
import torch
from transformers import AutoTokenizer, TextStreamer, GenerationConfig
from unsloth import FastLanguageModel
from IPython.display import clear_output
import time

# Check if GPU is available
print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"CUDA device: {torch.cuda.get_device_name(0)}")

# Load the Llama 3.2 model using Unsloth
def load_model():
    # Specify model name - choose a model size that will fit in Colab's GPU memory
    model_name = "unsloth/llama-3-8b-instruct"  # Using Unsloth's optimized version

    print(f"Loading model: {model_name}")

    # Load the model with Unsloth optimizations
    model, tokenizer = FastLanguageModel.from_pretrained(
        model_name=model_name,
        max_seq_length=4096,  # Choose appropriate sequence length
        dtype=torch.bfloat16,  # Use bfloat16 for faster inference
        load_in_4bit=True,     # Use 4-bit quantization to save memory
    )

    # Prepare the model for text generation
    model = FastLanguageModel.get_peft_model(
        model,
        target_modules=["q_proj", "k_proj", "v_proj", "o_proj",
                        "gate_proj", "up_proj", "down_proj"],
        r=16,  # LoRA attention dimension
        lora_alpha=32,
        lora_dropout=0,
    )

    clear_output()
    print("Model loaded successfully!")

    return model, tokenizer

# Define interview prompt function
def create_interview_prompt(resume, job_title, department, required_skills):
    prompt = f"""<s>[INST]
You are TechRecruit, an AI interviewer designed to conduct technical job interviews. You will be given a candidate's resume and must conduct a professional interview to evaluate their fit for a specific role. Your task includes analyzing the resume, asking relevant questions, evaluating responses, and providing a final assessment.

Instructions:
1. Resume Analysis: Carefully review the candidate's resume below to understand their background.
2. Interview Format:
   - Begin with a brief professional introduction
   - Ask 6-8 questions in total:
     * 2-3 questions about specific work experiences mentioned in their resume
     * 2-3 questions about technical skills or projects mentioned in their resume
     * 2 general questions related to the role they're applying for
   - End with a comprehensive evaluation
3. Evaluation Criteria:
   - Rate each answer on: Relevance (1-5), Technical Depth (1-5), Communication (1-5)
   - Provide a final score out of 100
   - Clearly state whether the candidate PASSED or FAILED the interview
   - Include a brief summary of strengths and areas for improvement

Resume:
{resume}

Role:
Position: {job_title}
Department: {department}
Required Skills: {required_skills}

Conduct the complete interview process following the instructions above. Start with your introduction, ask all questions, and provide your final evaluation.
[/INST]</s>
"""
    return prompt

# Sample resume
sample_resume = """
JOHN DOE
Software Engineer | john.doe@email.com | (123) 456-7890 | linkedin.com/in/johndoe | github.com/johndoe

EXPERIENCE
Senior Software Engineer, XYZ Tech Inc.
June 2021 - Present
- Led development of a microservices architecture that improved system scalability by 40%
- Implemented CI/CD pipelines reducing deployment time from 2 days to 2 hours
- Mentored junior developers and conducted code reviews

Software Developer, ABC Software Solutions
March 2018 - May 2021
- Developed RESTful APIs using Node.js and Express serving 50k daily requests
- Optimized database queries reducing response time by 30%
- Collaborated with cross-functional teams to deliver features on time

PROJECTS
E-commerce Platform (React, Node.js, MongoDB)
- Built a full-stack e-commerce application with user authentication, product catalog, and payment processing
- Implemented Redis caching to improve performance

Weather Prediction App (Python, TensorFlow)
- Created a machine learning model to predict weather patterns with 85% accuracy
- Deployed as a Flask API with a React frontend

SKILLS
Languages: JavaScript, Python, Java, SQL
Frameworks: React, Node.js, Express, Django, Flask
Tools: Git, Docker, Kubernetes, AWS, Jenkins
Databases: MongoDB, PostgreSQL, MySQL, Redis

EDUCATION
Bachelor of Science in Computer Science
University of Technology, 2014-2018
GPA: 3.8/4.0
"""

# Job details
job_title = "Senior Full Stack Developer"
department = "Engineering"
required_skills = "JavaScript, React, Node.js, MongoDB, AWS, CI/CD"

# Function to run the interview
def run_interview(model, tokenizer):
    # Create the interview prompt
    interview_prompt = create_interview_prompt(sample_resume, job_title, department, required_skills)

    # Set up generation parameters
    generation_config = GenerationConfig(
        max_new_tokens=2048,
        temperature=0.7,
        top_p=0.9,
        do_sample=True
    )

    # Generate the response with streaming
    print("Starting the interview process...\n")
    text_streamer = TextStreamer(tokenizer, skip_prompt=True)

    # Generate text with Unsloth's optimized generation
    start_time = time.time()

    with torch.no_grad():
        outputs = model.generate(
            **tokenizer(interview_prompt, return_tensors="pt").to("cuda"),
            generation_config=generation_config,
            streamer=text_streamer
        )

    end_time = time.time()
    print(f"\nGeneration time: {end_time - start_time:.2f} seconds")

    # Return the generated text
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Function to simulate a conversation
def simulate_conversation(model, tokenizer):
    print("Starting a conversation with Llama 3.2. Type 'exit' to end.")

    conversation_history = ""

    while True:
        user_input = input("\nYou: ")
        if user_input.lower() == 'exit':
            break

        # Update conversation history with user input
        conversation_history += f"<s>[INST] {user_input} [/INST]</s>"

        # Set up generation parameters
        generation_config = GenerationConfig(
            max_new_tokens=512,
            temperature=0.7,
            top_p=0.9,
            do_sample=True
        )

        # Generate response with streaming
        print("\nLlama 3.2: ", end="")
        text_streamer = TextStreamer(tokenizer, skip_prompt=True)

        with torch.no_grad():
            outputs = model.generate(
                **tokenizer(conversation_history, return_tensors="pt").to("cuda"),
                generation_config=generation_config,
                streamer=text_streamer
            )

        # Get the response and update conversation history
        response = tokenizer.decode(outputs[0], skip_special_tokens=True).split("[/INST]</s>")[-1].strip()
        conversation_history += response

# Main execution
def main():
    # Load the model
    model, tokenizer = load_model()

    print("Choose an option:")
    print("1. Run the technical interview")
    print("2. Start an interactive conversation")
    choice = input("Enter your choice (1 or 2): ")

    if choice == "1":
        run_interview(model, tokenizer)
    else:
        simulate_conversation(model, tokenizer)

# Call the main function when the script is run
main()
```
